membuat array

```python
nilai_mahasiswa = []
```

Input nilai mahasiswa

```python
for i in range(10):
    angka = int(input(f"Masukkan nilai mahasiswa{i+1}: "))
    nilai_mahasiswa.append(angka)
```

Nilai Mahasiswa

```python
print("Nilai Mahasiswa:", nilai_mahasiswa)
print("Nilai maksimum:", max(nilai_mahasiswa))
print("Nilai minimum:", min(nilai_mahasiswa))
print("Rata-rata:", sum(nilai_mahasiswa)/len(nilai_mahasiswa))
```

jumlah lulus

```python
jumlah_lulus = 0
for nilai in nilai_mahasiswa:
    if nilai >= 60:
        jumlah_lulus += 1
print("Jumlah mahasiswa yang lulus:", jumlah_lulus)
print("Jumlah mahasiswa yang tidak lulus", len(nilai_mahasiswa) - jumlah_lulus)
```

pembuatan grafik batang

```python
import matplotlib.pyplot as plt


label_grafik = ["Nilai Tertinggi", "Nilai Terendah"]
data_grafik = [max(nilai_mahasiswa), min(nilai_mahasiswa)]

plt.bar(label_grafik, data_grafik)
plt.title("Grafik Perbandingan Nilai")
plt.ylabel("Nilai")
plt.show()
```

Menambahkan grafik kelulusan

```python
import matplotlib.pyplot as plt

label_pie = ["Lulus", "Tidak Lulus"]
data_pie = [jumlah_lulus, len(nilai_mahasiswa)-jumlah_lulus]

plt.pie(data_pie, labels=label_pie, autopct='%1.1f%%')
plt.title("Grafik Persentase Kelulusan")
plt.axis('equal')
plt.show()
```
