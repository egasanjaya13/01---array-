membuat array

```python
nilai_mahasiswa = []
```

Input nilai mahasiswa

```python
for i in range(10):
    angka = int(input(f"Masukkan nilai mahasiswa{i+1}: "))
    nilai_mahasiswa.append(angka)
```

Nilai Mahasiswa

```python
print("Nilai Mahasiswa:", nilai_mahasiswa)
print("Nilai maksimum:", max(nilai_mahasiswa))
print("Nilai minimum:", min(nilai_mahasiswa))
print("Rata-rata:", sum(nilai_mahasiswa)/len(nilai_mahasiswa))
```

jumlah lulus

```python
jumlah_lulus = 0
for nilai in nilai_mahasiswa:
    if nilai >= 60:
        jumlah_lulus += 1
print("Jumlah mahasiswa yang lulus:", jumlah_lulus)
print("Jumlah total mahasiswa yang tidak lulus", len(nilai_mahasiswa) - jumlah_lulus)
```
